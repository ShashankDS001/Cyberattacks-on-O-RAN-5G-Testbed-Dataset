## 1. Flow ID

_Unique identifier for each network flow, typically combining source/destination IPs, ports, and protocols._

```python
# Count the unique Flow IDs
unique_flow_count = df['Flow ID'].nunique()
unique_flow_count
```

_Source IP address of the flow._

## 7. Timestamp

_Start time of the flow._

```python
# Convert Timestamp column to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
# Extract hour from Timestamp
df['Hour'] = df['Timestamp'].dt.hour

# Count the occurrences per hour
hourly_traffic = df['Hour'].value_counts().sort_index()

# Plot traffic distribution by hour

plt.figure(figsize=(10, 6))
hourly_traffic.plot(kind='bar', color='skyblue')
plt.title('Traffic Distribution by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Set Timestamp as index for time series analysis
df.set_index('Timestamp', inplace=True)

# Resample by hour and calculate the sum of traffic bytes
hourly_traffic_volume = df['Flow Byts/s'].resample('h').sum()

# Plot the traffic volume over time
plt.figure(figsize=(12, 6))
hourly_traffic_volume.plot()
plt.title('Traffic Volume Over Time (Hourly)')
plt.xlabel('Time')
plt.ylabel('Traffic Volume (Flow Byts/s)')
plt.show()
```
